```python
# Buyer overdrawal with scheduled non-zero when importing power
deviation.loc[
    (deviation['actual'] < 0) & 
    (deviation['scheduled'] < 0) &  # Additional condition
    (
        (deviation['scheduled'] != 0) & 
        (deviation['actual'] < deviation['scheduled']) & 
        (deviation['frequency'] < 49.95) & 
        (deviation['frequency'] > 49.90)
    ),
    'price'
] = deviation['deviation'] * 1.5 * 10 * deviation['Ncd_values']  # overdraw at abnormal freq

deviation.loc[
    (deviation['actual'] < 0) & 
    (deviation['scheduled'] < 0) &  # Additional condition
    (
        (deviation['scheduled'] != 0) & 
        (deviation['actual'] < deviation['scheduled']) & 
        (deviation['frequency'] > 50.03) & 
        (deviation['frequency'] < 50.05)
    ),
    'price'
] = deviation['deviation'] * 0.75 * 10 * deviation['Ncd_values']  # overdraw at 50.03 to 50.05 freq

deviation.loc[
    (deviation['actual'] < 0) & 
    (deviation['scheduled'] < 0) &  # Additional condition
    (
        (deviation['scheduled'] != 0) & 
        (deviation['actual'] < deviation['scheduled']) & 
        (deviation['frequency'] <= 49.90)
    ),
    'price'
] = deviation['deviation'] * 2 * 10 * deviation['Ncd_values']  # overdraw at f less than 49.90

deviation.loc[
    (deviation['actual'] < 0) & 
    (deviation['scheduled'] < 0) &  # Additional condition
    (
        (deviation['scheduled'] != 0) & 
        (deviation['actual'] < deviation['scheduled']) & 
        (deviation['frequency'] >= 50.05)
    ),
    'price'
] = 0  # With schedule non-zero at high frequency
```

```python
# Conditions overdraw at noraml freq schedule less than 100 to buy 
# Conditions
condition = (
    (deviation['actual'] < 0) & 
    (deviation['scheduled'] < 0) &  # Additional condition
    (
        (deviation['scheduled'] != 0) & 
        (deviation['actual'] < deviation['scheduled']) & 
        (deviation['frequency'] >= 49.95) & 
        (deviation['frequency'] <= 50.03) &  
        (deviation['scheduled'] >= -100) &
        (deviation['percent'] > 20)
    )
)

# Calculate minimum_term_4
minimum_term_4 = np.minimum(10, (-deviation['scheduled'] * 0.2))

# Calculate price_1
price_1 = minimum_term_4 * 10 * deviation['Ncd_values']

# Calculate price_2
price_2 = -((deviation['deviation'] + minimum_term_4) * 1.2 * 10 * deviation['Ncd_values'])

# Calculate final price as the negative sum of price_1 and price_2
price = -(price_1 + price_2)

# Assign the calculated values to the 'price' column where condition is True
deviation.loc[condition, 'price'] = price




# Conditions for percent less than or equal to 20
condition_less_than_or_equal_to_20 = (
    (deviation['actual'] < 0) & 
    (deviation['scheduled'] < 0) &  
    (deviation['scheduled'] != 0) & 
    (deviation['actual'] < deviation['scheduled']) & 
    (deviation['frequency'] >= 49.95) & 
    (deviation['frequency'] <= 50.03) & 
    (deviation['scheduled'] >= -100) &   # added due to Haryana case
    (deviation['percent'] <= 20)
)

# Calculate minimum value and price
minimum_value_5 = np.minimum(10, -deviation['deviation'])
price = (
    minimum_value_5 * 10 * deviation['Ncd_values'] - 
    (deviation['deviation'] + minimum_value_5) * 1.2 * 10 * deviation['Ncd_values']
)

# Assign the calculated values to the 'price' column where condition is True
deviation.loc[condition_less_than_or_equal_to_20, 'price'] = -price


```
